```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('meteonogal.csv', sep=',', engine='python')
df.head()
```

```python
df.dtypes
```

```python
df.head()
```

```python
# para eliminar columnas sin informaciÃ³n, se buscÃ³ la manera
# de eliminar aquellas que no estÃ©n nombradas 'unnamed'
df = df.drop(df.columns[df.columns.str.contains('unnamed:',case = False)],axis = 1)
df.head()
```

```python
df.dtypes
```

```python
# se quita el primer renglÃ³n donde se muestran las unidades
df = df.drop([0], axis=0)
df.head(3)
```

```python
df[df.columns[2:16]] = df[df.columns[2:16]].astype(float)
df.dtypes
```

```python
corr = df.corr()
corr
```

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Utilizando Seaborn

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# utilizando seaborn
#usando la misma mÃ¡scara superior
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```
